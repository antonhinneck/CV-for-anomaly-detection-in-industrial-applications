```python
# выбрем необходимый датасет из юры списка
dataset = 3
data = data_init[data_init.pipeline_num == dataset].drop(['pipeline_num','Unnamed: 0','Unnamed: 0.1'],axis=1)
jou = jou_init[jou_init.pipeline_num == dataset].drop(['pipeline_num','Unnamed: 0'],axis=1)
faults = faults_init[faults_init.pipeline_num == dataset].drop(['pipeline_num','Unnamed: 0'],axis=1)
```

```python
shov = jou['Местоположение начала секции по дистанции трубопровода, м'].values.astype(float)
print('len=',len(shov))
print('max dist=',max(shov))
```

```python
# очистка, процент информации который останется
print((len(data.dropna()))/len(data)*100)
print((len(data.dropna(axis=1)))/len(data)*100)
```

```python
data=data.dropna(axis=1)
data.index=data.Dist
data = data.drop(['Dist','Dist', 'Time', 'work_flag'],axis=1)
print(data.columns)
print('\n\n')
print('Diff Distanse',data.index[-1] - data.index[0] - shov[-1] + shov[0])
```

```python
def view(data,mi,ma):
    return data[(data.index >= mi) & (data.index <= ma)]
```

```python
print(shov[100:120])
tt = (view(data,600,1500) -view(data,600,1500).median()).sum(axis=1)
print(tt)
ttt= tt.copy()*(-1)
ttt.index = list(map( lambda x: round(x,3) , tt.index))
shs = shov[100:120]
shs
```

```python
tttt = pd.DataFrame()
tttt['ndex']=ttt.index
tttt['value']=ttt.values
delta = np.zeros(len(ttt))
for i in shs:
    idx = tttt.ndex[tttt.ndex > i].index[0]
    delta+=signal.unit_impulse(len(delta), range(idx-50,idx+50))
# plt.plot(delta[50000:150000])
```

```python
delta = delta *0.5
```

```python
scl = MinMaxScaler()
ttt_ = scl.fit_transform(ttt.values.reshape(-1, 1)).ravel()
```

```python
ttt_[ttt_ < 0.2] = 0
```

```python
plt.plot(ttt_)
plt.plot(delta)
```

```python
plt.plot(np.correlate(ttt_, delta, mode='full'))
zz = np.argmax(np.correlate(ttt_, delta, mode='full'))-len(ttt)
print(zz)
```
