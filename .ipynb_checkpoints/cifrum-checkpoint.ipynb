```python
# выбрем необходимый датасет из юры списка
dataset = 5
data = data_init[data_init.pipeline_num == dataset].drop(['pipeline_num','Unnamed: 0','Unnamed: 0.1'],axis=1)
jou = jou_init[jou_init.pipeline_num == dataset].drop(['pipeline_num','Unnamed: 0'],axis=1)
faults = faults_init[faults_init.pipeline_num == dataset].drop(['pipeline_num','Unnamed: 0'],axis=1)
```

```python
shov = jou['Местоположение начала секции по дистанции трубопровода, м'].values.astype(float)
print('len=',len(shov))
print('max dist=',max(shov))
```

```python
# очистка, процент информации который останется
print((len(data.dropna()))/len(data)*100)
print((len(data.dropna(axis=1)))/len(data)*100)
```

```python
data=data.dropna(axis=0)
data.index=data.Dist
data = data.drop(['Dist','Dist', 'Time', 'work_flag'],axis=1)
print(data.columns)
print('\n\n')
print('Diff Distanse',data.index[-1] - data.index[0] - shov[-1] + shov[0])
```

```python
def view(data,mi,ma):
    return data[(data.index >= mi) & (data.index <= ma)]
```
